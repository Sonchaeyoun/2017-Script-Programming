### 2번 문제

* 다형성에 대해 설명하고 다형성을 보여주는 자신만의 파이썬 코드 예제를 제시하시오.

다형성은 다른 클래스들의 인스턴스들이 상속 관계 내에 있다면 같은 멤버 함수 호출에 대해 각각 다르게 반응하도록 해 주는 기능이다. 이는 코드의 가독성을 높여주고, 작성할 코드의 양이 줄어들고, 적은 코딩으로 다양한 객체들에게 유사한 작업을 수행시킬 수 있다는 장점이 있다. 특히 형 선언이 없는 파이썬에서는 이를 적용하기가 더욱 용이하다. 객체의 형이 실시간으로 결정되므로 하나의 메소드에 의해 처리될 수 있는 객체의 종류에 제한이 없어지는 것이다. 다음은 다형성을 보여주는 나만의 파이썬 코드 예제이다.

```python
class Food:
    def flavor(self):
        print('delicious')
        
class Cookie(Food):
    def flavor(self):
        print('sweet')
        
class Kimchi(Food):
    def flavor(self):
        print('salty')
        
class Lemon(Food):
    pass

for each in (Cookie(), Kimchi(), Lemon()):
    each.flavor()
```

위 예제를 통해 상속 관계 내에 있는 다른 클래스들의 인스턴스들이 같은 멤버 함수 호출에 대해 각각 다르게 반응한다는 것을 직접 눈으로 확인할 수 있다.

### 3번 문제

* 다음 각 요구사항 모두를 만족시키는 Counter 클래스를 코딩하시오 (정답을 각 요구사항별로 입력할 필요 없이 3번 문제에 대해 1개의 클래스 정의 코드를 제시하면 된다.)

* 요구사항 1. 생성자에 count 값과 step 값을 인자로 받을 수 있다.
    * count: Counter 인스턴스가 지니는 초기 정수 값
    * step: Counter 인스턴스의 count가 증가되는 증분 (defalt 값: 1)
  
        * c = Counter(10)<br/>

* 요구사항 2. 다음과 같이 Counter의 인스턴스를 출력을 해주는 __str__() 메소드를 Counter 클래스 내에 구현하시오.

          [Count (step: 1)] 10 
        * print(d)<br/> 
          [Count (step: 2)] 10 

* 요구사항 3. 다음과 같이 step에 주어진 증분만큼 count를 증가시키는 incr() 메소드를 Counter 클래스 내에 구현하시오.

        * c.incr()

        * print(c)<br/> 
          [Count (step: 1)] 11 
        * print(d)<br/> 
          [Count (step: 2)] 12 

* 요구사항 4. Counter 클래스 내에 관련 메소드를 추가하여 인스턴스 객체를 직접 호출(call)할 수 있도록 하시오. 인스턴스 객체를 직접 호출했을 때에 내부적으로 incr() 메소드를 호출하는 방법으로 구현하시오.

        * c() 
        * d() 
        * print(c)<br/> 
          [Count (step: 1)] 12 
        * print(d)<br/> 
          [Count (step: 2)] 14 

* 요구사항 5. 다음과 같은 두 개의 산술 연산 (+, -)이 수행될 수 있도록 Counter 클래스 내에 관련 메소드를 추가하시오.

        * c = c + 5 
        * d = d - 5 
        * print(c)<br/> 
          [Count (step: 1)] 17 
        * print(d)<br/> 
          [Count (step: 2)] 9 

* 요구사항 6. 다음과 같은 관계연산 (>, <, ==)이 수행될 수 있도록 Counter 클래스 내에 관련 메소드를 추가하시오.

        * print(c) > 10<br/> 
          True 
        * print(d) > 10<br/> 
          False 
        * print(c) < 10<br/> 
          False 
        * print(d) < 10<br/> 
          True 
        * print(c) == 17<br/> 
          True 
        * print(d) != 9<br/> 
          False 

```python
class Counter():
    
    def __init__(self, count=1, step=1):
        self.count = count
        self.step = step
    
    def __str__(self):
        return "[Count (step: %d)] %d" % (self.step, self.count)
    
    def incr(self):
        self.count += self.step
        
    def __call__(self):
        self.incr()
        
    def __add__(self, n):
        self.count += n
        return self.count
    
    def __sub__(self, n):
        self.count -= n
        return self.count
    
    def __cn1__(self, n):
        return self.count < n
    
    def __cn2__(self, n):
        return self.count > n
    
    def __cns__(self, n):
        return self.count == n
    
    def __cnd__(self, n):
        return self.count != n
    
# 확인하는 코드
c = Counter(10)
d = Counter(10, 2)
print(c)
print(d)
print('\n')
c.incr()
d.incr()
print(c)
print(d)
print('\n')
c()
d()
print(c)
print(d)
print('\n')
c = c + 5
print(c)
d = d - 5
print(d)
print('\n')
print(c > 10)
print(d > 10)
print(c < 10)
print(d < 10)
print(c == 17)
print(d != 9)
```

요구사항에 따라 Counter 클래스에 생성자, 출력 메소드, 증가 메소드, 호출 메소드, 산술 연산 메소드, 관계 연산 메소드들을 넣어 코딩하였다. 그리고 마지막에는 이를 확인하는 코드를 넣어 나의 코드가 맞는지 확인해보았다.

### 4번 문제

* 다음은 내장 자료형 list를 서브클래싱하여 만든 MySet 클래스 정의 내용이다. 다음 클래스 정의에서 __init__(), __str()__(), elimicate_duplicate()의 세 개의 메소드 코드 내용을 자신이 다른 사람에게 가르친다고 생각하며 설명해보시오.

    * MySet은 집합(Set) 자료형을 정의하려는 의도하에 만들어진 클래스이다.

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)

    
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print(s)
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print(t)
```

우선 init()은 리스트를 매개 변수로 받아서 그의 각 원소들을 self로 추가한다. 그 후 self 리스트를 eliminate_duplicate()함수에 인자로 보낸다. str()은 result 변수에 self 리스트에 담겨진 문자를 넣고, 필요없는 것들을 없애기 위해 result = result[0:len(result)-2] + "}"를 해 준다. 그리고 이렇게 만들어진 result의 값을 반환한다. 마지막으로 eliminate_duplicate()는 self 리스트를 인자로 받아 이중 for문을 사용하여 기존에 있던 것인지 확인한다. 기존에 있던 것이 아니면 s에 append()를 사용해 저장한다. 그리고 이를 다시 빈 리스트 l에 저장한다. 즉, 중복된 원소가 생기지 않게 하려는 것이다.

    * 요구사항 2. & 연산으로 두 집합의 교집합을 반환한다.

        * t = MySet([3, 4, 5]) 
        * u = s - t

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)
            
    def __or__(set1, set2): # set집합을 사용해 | 연산으로 두 집합의 합집합을 반환하게 했다.
        set1 = set(set1)
        set2 = set(set2)
        set1 = set1|set2
        set1 = list(set1)
        res = "MySet: {" 
        for x in set1:
            res = res + str(x) + " ,"
        res = res[0:len(res)-2] + "}"
        return res
    
    def __and__(set1, set2): # set집합을 사용해 & 연산으로 두 집합의 교집합을 반환하게 했다.
        set1 = set(set1)
        set2 = set(set2)
        set1 = set1&set2
        set1 = list(set1)
        res = "MySet: {"
        for x in set1:
            res = res + str(x) + " ,"
        res = res[0:len(res)-2] + "}"
        return res

    def __sub__(set1, set2): # set집합을 사용해 - 연산으로 두 집합의 차집합을 반환하게 했다.
        set1 = set(set1)
        set2 = set(set2)
        set1 = set1-set2
        set1 = list(set1)
        res = "MySet: {"
        for x in set1:
            res = res + str(x) + " ,"
        res = res[0:len(res)-2] + "}"
        return res
    
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print(s)
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print(t)
    u = s | t 
    print(u)
    u = s & t 
    print(u)
    s = MySet([1, 2, 3]) 
    t = MySet([3, 4, 5]) 
    u = s - t 
    print(u)
```

문제에서 제시한 요구사항 별로 메소드를 만들어 코딩하였다. 각 메소드는 모두 연산을 달리했다는 차이점만 제외하면 인자로 받은 set1과 set2의 연산 결과를 set1에 넣고 이를 다시 res 변수에 넣어 반환하는 것은 똑같다. 그리고 이를 확인하기 위해 테스트하는 코드를 넣었다.
